Problem6. Write a program to swap values of two numbers entered by the user.

```python
val_one = 15
val_two = 100
print(f"Before swap val_one: {val_one}")
print(f"Before swap val_two: {val_two}")
tmp = val_one
val_one = val_two
val_two = tmp
print(f"After swap val_one: {val_one}")
print(f"After swap val_two: {val_two}")
```
